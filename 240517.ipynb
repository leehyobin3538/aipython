```python
dataset = pd.DataFrame({
'x': [1, 2, 3, 4, 5],
'y': [10, 20, 15, 25, 30],
'category': ['A', 'B', 'A', 'B', 'A']
})

sns.scatterplot(x='x',y='y',hue='category',data=dataset)
plt.show()
```

```python
dataset2 = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})

sns.countplot(x='category',hue='category',data=dataset2)
plt.show()
```
