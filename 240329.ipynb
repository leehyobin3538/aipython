1.**분을 입력 받아서 시간과 분을 출력하시오** (.format 메소드 활용)

```python
# 나의 풀이
A=int(input())
print("{}시간 {}분".format(A//60,A%60))
```

```python
# 교수님 풀이
in_m=int(input("분을 입력하시오"))
h=in_m//60
m=in_m%60
print("{:d} -> {:d}시간 {:d}분".format(in_m,h,m))
```

2. **섭씨 온도를 입력 받아서 화씨 온도를 출력하시오** (% 연산자 활용)

```python
# 나의 풀이
A=int(input())
B=A*1.8+32
print("섭씨 %d도는 화씨 %d도입니다."%(A,B))
```

```python
# 교수님 풀이
c_t=int(input("섭씨 온도를 입력하시오"))
f_t=c_t*1.8+32
print("섭씨 %d도는 화씨 %.1f도입니다." % (c_t,f_t))
```

3. **원의 반지름을 입력 받아서 원의 둘레와 면적을 출력하시오** (PI=3.14)

```python
# 나의 풀이
A=int(input())
PI=3.14
B=A*3.14*2
C=A**2*3.14
print("반지름이 {}인 원의 둘레는 {}이고 면적은 {}입니다.".format(A,B,C))
```

```python
# 교수님 풀이

# PI는 상수
PI=3.14
r=int(input("원의 반지름을 입력하시오"))
circumference=2*PI*r
area=PI*(r**2)
print("반지름이 {}인 원의 둘레 {}이고 면적은 {}입니다.".format(r,circumference,area))
```
