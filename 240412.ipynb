```python
#함수의 정의 및 실행
def add(x,y):
  s=x+y
  a=s/2
  return s,a

h,b=add(10,20)
print("합: ",h,"평균: ",b)
```

```python
#다음 리스트의 모든 요소를 제곱하여 새로운 리스트를 생성하세요(사용자 함수 정의)


def square(numbers):
  newlst=[]
  for num in numbers:
    newlst.append(num**2)
  return newlst

new=square([1,2,3,4,5])
print(new)
```

```python
#bmi 지수를 측정하는 함수
#bmi=체중(kg)/{신장(m)}**2


#단위주의

def BMI(weight,height):
    result=weight/(0.1*height)**2
    return result

print(BMI(44,160))
```
