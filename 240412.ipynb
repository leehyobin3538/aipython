함수: 함수는 독립적으로 호출될 수 있습니다. 함수를 호출할 때는 함수의 이름을 사용하고 필요한 매개변수를 전달합니다.
메소드: 메소드는 해당 클래스나 객체에 속한 함수이기 때문에 해당 클래스나 객체의 이름을 사용하여 호출됩니다.

```python
#다음 리스트의 모든 요소를 제곱하여 새로운 리스트를 생성하세요
numbers=[1,2,3,4,5]
newlst=[]
for num in numbers:
  newlst.append(num**2)
print(newlst)
```
