```python
a=int(input("값을 입력하시오"))
if a%2==0:
  print("a는 짝수입니다")
elif a%2==1:
  print("a는 홀수입니다")
```

```python
# 1~10까지의 짝수, 홀수 누적합

even_tot=0
odd_tot=0

for i in range(1,11):
  if i%2==0:
    even_tot+=i
  else:
    odd_tot+=i

print(even_tot)
print(odd_tot)

```
